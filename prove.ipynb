```python
import os 
print(os.getcwd())
os.chdir('/data/francesca/lbae')

import new_storage
from new_storage import LipidImage, MaldiData
import shelve
meta_path = "new_data/metadata"
with shelve.open(meta_path) as db:
    brain_info = db.get("brain_info", {})
    print("Stored Brain IDs:", list(brain_info.keys()))
    # Optionally, check slices for a given brain
    if "Female1" in brain_info:
        print("Slices for Female1:", list(brain_info["Female1"].keys()))


storage = MaldiData(path_db="new_data/")


# Now you can retrieve a lipid image
lipid_data = storage.get_lipid_image("Female1", 74.0, "HexCer 42:2;O2")
if lipid_data:
    print("Successfully stored and retrieved scatter data!")
    print(f"Number of points: {len(lipid_data.image)}")
    
    # Create a scatter plot to verify<
    
    import matplotlib.pyplot as plt
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(
        lipid_data.image[:, 0],  # x coordinates
        lipid_data.image[:, 1],  # y coordinates
        c=lipid_data.image[:, 2],  # values for color
        cmap="hot"
    )
    plt.colorbar(scatter, label="Lipid Value")
    plt.title(f"Scatter Plot of {lipid_data.name}")
    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.show()
else:
    print("Lipid image not found in the database.")
```
